```python
#extract data - dairy
def getDairy(k):
    
    #men k=7
    #women k=10

    #Fluid milk total"
    i=j=108
    milk = df.ix[i:j,k]

    # "Butter", "Cheese", "Yogurt", "Dairy, Other".
    i=113
    j=116
    others = df.ix[i:j,k]

    milkMatrix = milk.as_matrix()
    othersMatrix = others.as_matrix()

    allMatrix = np.concatenate((milkMatrix,othersMatrix),axis=0)
   
    #print('Values: ' + str(allMatrix))
    #allMatrix = np.reshape(allMatrix, (9, 1))
    #print('Values: ' + str(allMatrix))
    return allMatrix
```

```python
#plot - dairy
def plotDairy(gender,caption):
    years = ['94-98','03-04','05-06','07-08']
    yearsNum = np.array([1,2,3,4])
    plt.xticks(yearsNum, years,fontsize=20)
    plt.title('Dairy types over time (' + gender +')',fontsize=20) 
    plt.ylabel('Mean %',fontsize=20) 
    plt.xlabel('Years',fontsize=20)
    plt.plot(yearsNum,finalMatrix[0,0],marker='o',label='Fluid milk total')
    plt.plot(yearsNum,finalMatrix[0,1],marker='o',label='Butter')
    plt.plot(yearsNum,finalMatrix[0,2],marker='o',label='Cheese')
    plt.plot(yearsNum,finalMatrix[0,3],marker='o',label='Yogurt')
    plt.plot(yearsNum,finalMatrix[0,4],marker='o',label='Dairy, other')

    plt.rcParams['figure.figsize'] = (12.0,10.0) 
    plt.annotate(caption, xy=(0.1, -0.2), xycoords='axes fraction')
    plt.rcParams.update({'font.size': 12})
    plt.legend()
    plt.show()
```
